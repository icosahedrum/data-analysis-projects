## Clean Your Data

In the last lesson, we cleaned the data related to San Francisco. Pull up your notebook from the last lesson and use it as a reference to clean up these new dataframes.

```python
# Clean your data here!
# data["Type"] = data["Type"].fillna("Conventional") # to clean type
#
# data = data.drop(["Grade"], axis=1)
# data = data.drop(["Environment"], axis=1)
# data = data.drop(["Quality"], axis=1)
# data = data.drop(["Condition"], axis=1)
# data = data.drop(["Appearance"], axis=1)
# data = data.drop(["Storage"], axis=1)
# data = data.drop(["Crop"], axis=1)
# data = data.drop(["Trans Mode"], axis=1)

def clean_produce_df(df: pd.DataFrame) -> pd.DataFrame:
    cols_to_drop = [
        "Grade",
        "Environment",
        "Quality",
        "Condition",
        "Appearance",
        "Storage",
        "Crop",
        "Trans Mode"
    ]


    data = df.copy()


    if "Type" in data.columns:
        data["Type"] = data["Type"].fillna("Conventional")

    existing_cols = [col for col in cols_to_drop if col in data.columns]
    data = data.drop(existing_cols, axis=1)

    return data

philly_cleaned = clean_produce_df(philly)
balmer_cleaned = clean_produce_df(balmer)
boston_cleaned = clean_produce_df(boston)
ny_cleaned = clean_produce_df(ny)
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
northeast = pd.concat([philly_cleaned, balmer_cleaned, boston_cleaned, ny_cleaned], ignore_index=True)
```
