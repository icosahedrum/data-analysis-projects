### Creating Visualizations

Now that the data is ready, you can use the dataset to create a number of different visualizations. You do not need to make the visualizations perfect; this is just for practice.

#### Line Chart

```python
# Single line chart
# df.plot(y = 'danceability_%', x = 'released_year')

average_danceability_by_year = df.groupby('released_year')['danceability_%'].mean()

plt.figure(figsize=(10, 6))
average_danceability_by_year.plot(kind='line', marker='o')
plt.title('Average Danceability Percentage by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average Danceability %')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Multiple Line chart 
df.plot(y = 'danceability_%', x = 'released_year')
```

#### Bar Chart

```python
# Vertical bar chart
average_bpm_by_year = df.groupby('released_year')['bpm'].mean()

plt.figure(figsize=(10, 6))
average_bpm_by_year.plot(kind='bar')
plt.title('Average BPM by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average BPM per Song')
plt.grid(True)
plt.show()
```

```python
# Horizontal bar chart
average_bpm_by_year = df.groupby('released_year')['bpm'].mean()

plt.figure(figsize=(10, 6))
average_bpm_by_year.plot(kind='barh')
plt.title('Average BPM by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average BPM per Song')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg = df.groupby('released_year')[['bpm', 'danceability_%']].mean().sort_index()

x = np.arange(len(avg.index))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width/2, avg['bpm'], width, label='Average BPM')
ax.bar(x + width/2, avg['danceability_%'], width, label='Danceability (%)')


ax.set_title('Average BPM and Danceability (%) by Release Year')
ax.set_xlabel('Release Year')
ax.set_ylabel('Average Value')
ax.set_xticks(x)
ax.set_xticklabels(avg.index, rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', linewidth=0.5)
ax.legend()

plt.tight_layout()
plt.show()
```
